## Parte 1

Consideremos un sistema en el que existe una característica 'x' y una variable dependiente 'y' que se relacionan como:

$y = a * x$ 

Supongamos también que para cada valor fijo de 'x' la medida 'y' viene dada por una pdf gaussiana centrada en 

$a * x$

y con sigma $\sigma$. 

Escribe una funcion de R que tome como input un vector aleatorio de N elementos de 'x' que tome valores entre [0, 10], el factor 'a', y sigma; y genere el correspondiente vector 'y'.

```r
get.y <- function(x, angular.coeff, std.dev){
    my.output <- c()
    for (values in x){
    my.output <- c(my.output, rnorm(1, angular.coeff*values, std.dev))
    }
    return(my.output)
    }
```

## Parte 2

Una vez terminada la función, genera el vector x con N = 100, y un vector y con a = 2 y sigma = 0.2. Pintalos en una grafica. 

```r
# Definimos los parametros
N <- 100
a <- 2
sigma <- 0.2

# Generamos los valores de x ...
my.x <- runif(N,0,10)

# ... y los valores de y
my.y <- get.y(my.x,a,sigma)
```

```r
plot(my.x,my.y)
```

## Parte 3

Escribe una función en R a la que le pases un vector 'y' dependiente, un vector 'x' con la caracteristica, y un parametro 'a' y te devuelva el valor de la loss function para ese valor del parámetro 'a'.

```r
# La función de loss es la distancia euclidea al cuadrado
# entre los valores y generados y el valor de la recta en ese punto
get.loss.func <- function(x.vec, y.vec, ang.coeff){
    f.x <- ang.coeff * x.vec
    loss <- (f.x - my.y)**2
    return(sum(loss))
}
```

## Parte 4

Llama a la función anterior con diferentes valores del parámetro 'a', y pinta la curva de la loss function en función de 'a'. 

Comprueba visualmente que el mínimo está en la posición que nos esperamos.

```r
loss.functions <- c()

for(i in seq(0, 5, by=0.1)){
    loss.functions <- c(loss.functions,  get.loss.func(my.x,my.y,i))
}
plot(seq(0, 5, by=0.1),loss.functions,
    xlab = "Coeficiente angular de la recta")
    ylab = "Función de Loss"
```

Lo que esperamos es que el minimo de la función de Loss se encuentre en correspondecia del valor del coeficiente angular de la recta usado para generar los valores del vector y, en nuestro caso: coeff. angular = 2.

Esto es exactamente lo que se puede observar en la grafica anterior.
